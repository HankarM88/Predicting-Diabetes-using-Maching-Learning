### Data Overview

The diabetes dataset, commonly referred to as the Pima Indians Diabetes Database, originates from the National Institute of Diabetes and Digestive and Kidney Diseases. It is widely used for binary classification tasks, where the goal is to predict whether a patient has diabetes based on diagnostic health measurements.ieving reliable performance.seving reliable performance.

**Features**:
- Pregnancies: Number of times the patient was pregnant.
- Glucose: Plasma glucose concentration from an oral glucose tolerance test.
- BloodPressure: Diastolic blood pressure (mm Hg).
- SkinThickness: Triceps skinfold thickness (mm).
- Insulin: Serum insulin concentration (mu U/ml).
- BMI: Body mass index, defined as weight (kg) / height (m)^2.
- DiabetesPedigreeFunction: A function scoring the likelihood of diabetes based on family history.
- Age: Age of the patient in years.
- Outcome: Binary label indicating whether the patient has diabetes (1) or not (0)

**Objectives**:
- Use these features to build a machine learning model to predict whether a person is at high risk of having diabetes.
- Determine the most crticial factors that will increase the risk of getting diabetes.

### Model Developpement

#### Sacling Data

```python
# Input features 
X = data.drop('Outcome',axis=1)
# Label
y = data["Outcome"]
# Split into train and test set 
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.30, random_state=42)
scaler = StandardScaler()
X_train_scaled = scaler.fit_transform(X_train)
X_test_scaled = scaler.fit_transform(X_test)
```

#### Training ML Models 

```python
import joblib
# Models dictionary  
model_dict = {'lr':LogisticRegression(),
              'knn': KNeighborsClassifier(n_neighbors=5,weights='uniform'),
              'dt' : DecisionTreeClassifier(criterion='gini', splitter='best', max_depth=10),
               'rfc': RandomForestClassifier(n_estimators = 100, max_depth =10, random_state = 43),
               'gbc': GradientBoostingClassifier(n_estimators=100, max_depth=6,learning_rate=0.1),
               'svc' : SVC(C=1.0,degree=3, kernel='rbf')}                                  

# Model names 
models = [key.upper() for key in model_dict.keys()]
acc_scores = []
f1_scores  = []
# Begin Training models 
for name, model in model_dict.items():
    print("="*40)
    print(f'Training {name.upper()} Model')
    model.fit(X_train_scaled, y_train)
    preds = model.predict(X_test_scaled)
    acc = accuracy_score(y_test, preds)
    f1  = f1_score(y_test, preds)
    acc_scores.append(acc)
    f1_scores.append(f1)
    joblib.dump(model, f"models/{name}.pkl") 
    print('Done.')
    
# Save the results in a dataframe 
results = pd.DataFrame(list(zip(models,acc_scores,f1_scores)),columns=['Model','Accuracy Score','F1 Score'])
# Show the results for comparaison 
results
```

```python
# Optimize the best perfoming model
rfc = RandomForestClassifier(random_state = 101)
params_grid = { 
    'n_estimators': [100, 300],
    'max_features': ['auto', 'sqrt', 'log2'],
    'max_depth' : [4, 5, 6, 7, 8, 10],
    'criterion' :['gini', 'entropy']
}
# Hyperparamter tuning 
cv_rfc = GridSearchCV(estimator=rfc, param_grid=params_grid, cv= 5)
cv_rfc.fit(X_train_scaled, y_train)
# Show the best estimator 
print(cv_rfc.best_params_)
```

```python
# Train the best classifier 
rfc = RandomForestClassifier(criterion='entropy', max_depth= 8, max_features= 'log2', n_estimators= 100)
rfc.fit(X_train_scaled, y_train)
rfc_preds = rfc.predict(X_test_scaled)
```

```python
# Show the Results 
print('\n')
print('Accuracy:', accuracy_score(y_test,rfc_preds))
print('\n')
print(classification_report(y_test,rfc_preds))
```

#### Feature Importance

```python
# Calculate feature importances
importances = rfc.feature_importances_
# Sort feature importances in descending order
indices = np.argsort(importances)[::-1]

# Rearrange feature names so they match the sorted feature importances
names = [data.columns[i] for i in indices]

# Barplot: Add bars
plt.bar(range(X.shape[1]), importances[indices])
# Add feature names as x-axis labels
plt.xticks(range(X.shape[1]), names, rotation=20, fontsize = 8)
# Create plot title
plt.title("Feature Importance")
# Show plot
plt.show()
```

#### Conclusion
The most critical factors that increases the risk of having diabetes are: blood Glucose level, body mass index and age.
